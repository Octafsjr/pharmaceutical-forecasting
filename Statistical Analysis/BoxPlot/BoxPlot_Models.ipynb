```python
import pandas as pd
import matplotlib.pyplot as plt

# Load residuals
arima = pd.read_csv("ARIMA_residuals.csv")
hw = pd.read_csv("HW_residuals.csv")
sarima = pd.read_csv("SARIMA_residuals.csv")
mlp = pd.read_csv("MLP_residuals.csv")
rnn = pd.read_csv("RNN_residuals.csv")
gru = pd.read_csv("GRU_residuals.csv")


# Organize residual data
residuals = [
    arima["residual_ARIMA"],
    hw["residual_HW"],
    sarima["residual_SARIMA"],
    mlp["residual_MLP"],
    rnn["residual_RNN"],
    gru["residual_GRU"]
]

# Comparative boxplot
plt.figure(figsize=(8,5))

plt.boxplot(
    residuals,
    labels=["ARIMA", "Holt-Winters", "SARIMA", "MLP", "RNN", "GRU"],
    patch_artist=True,
    boxprops=dict(facecolor="lightblue"),
    medianprops=dict(color="red", linewidth=2)
)

# Zero reference line
plt.axhline(
    y=0,
    color="black",
    linestyle="--",
    alpha=0.7
)

plt.title("Residuals Comparison Across Forecasting Models")
plt.xlabel("Forecasting Model")
plt.ylabel("Residuals")

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.6
)

plt.show()
```
