```python
import pandas as pd
from scipy.stats import friedmanchisquare


# ---------------------------------
# Ler arquivos CSV
# ---------------------------------

arima = pd.read_csv("ARIMA_results.csv")
hw = pd.read_csv("HoltWinters_results.csv")
sarima = pd.read_csv("SARIMA_results.csv")
rnn = pd.read_csv("RNN_results.csv")
mlp = pd.read_csv("MLP_results.csv")
gru = pd.read_csv("GRU_results.csv")

# ---------------------------------
# Teste Friedman
# ---------------------------------

def friedman_test(metric):

    stat, p = friedmanchisquare(

        arima[metric],
        hw[metric],
        sarima[metric],
        rnn[metric],
        mlp[metric],
        gru[metric]

    )

    print("\n===================================")
    print(f"Teste Friedman - {metric}")
    print("===================================")

    print(f"Estatística = {stat:.4f}")
    print(f"p-value = {p:.6f}")

    if p < 0.05:
        print(
            "Há diferença estatisticamente significativa entre os modelos."
        )
    else:
        print(
            "Não há diferença estatisticamente significativa entre os modelos."
        )

# ---------------------------------
# Executar para todas as métricas
# ---------------------------------

metrics = [
    "RMSE",
    "MAE",
    "MASE",
    "SMAPE"
]

for metric in metrics:
    friedman_test(metric)
```

```python
import scikit_posthocs as sp

# ---------------------------------
# Teste Conover-Friedman
# ---------------------------------

def conover_test(metric):

    # tabela no formato exigido pelo scikit-posthocs

    dados = pd.DataFrame({

        "ARIMA": arima[metric].values,
        "HW": hw[metric].values,
        "SARIMA": sarima[metric].values,
        "RNN": rnn[metric].values,
        "MLP": mlp[metric].values,
        "GRU": gru[metric].values

    })

    resultado = sp.posthoc_conover_friedman(
        dados,
        p_adjust="holm"
    )

    print("\n===================================")
    print(f"Conover pós-hoc - {metric}")
    print("===================================")

    print(resultado.round(5))

    return resultado

# ---------------------------------
# Executar para todas as métricas
# ---------------------------------

metrics = [
    "RMSE",
    "MAE",
    "MASE",
    "SMAPE"
]

conover_results = {}

for metric in metrics:

    conover_results[metric] = conover_test(metric)
```
